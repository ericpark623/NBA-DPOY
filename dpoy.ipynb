```python
shots = pd.read_csv('/Users/ericpark/Desktop/NBA_Player_Box_Scores.csv')
```

```python
shots = shots.dropna(axis = 1)
shots
```

For some reason, when I web scraped from nba.com, the column names changed, so I am changing it back to the proper names

```python
shots = shots[['Player','Team','Game Date','Column 5','FTM','FTA','FT%']]
shots = shots.rename(columns={'Game Date':'GP','Column 5':'MIN','FTM':'Contested 2PT Shots','FTA':'Contested 3PT Shots', 'FT%': 'Contested Shots'})
```

Must play 70% of games to qualify for league leaders in stats (58 games)
Less than 20 minutes is too low of a sample size

```python
shots['Total Contested Shots'] = (shots['Contested Shots']*shots['GP']).astype(int)
shots = shots[(shots['MIN']>20) & (shots['GP']>58)]
```

```python
shots
```

This dataset has diff% so I'm going to merge this dataset with the previous one

```python
defensive_rating = pd.read_csv('/Users/ericpark/Desktop/defensive_rating.csv')
```

```python
defensive_rating = defensive_rating[['Player','Column 5','Points','FGM','FGA','FG%','3PT']].dropna()
defensive_rating = defensive_rating.rename(columns={'Column 5':'GP','Points':'DFGM','FGM':'DFGA','FGA':'DFG%','3PT':'DIFF%'})
```

```python
defensive_rating = defensive_rating[defensive_rating['GP']>58]
avg_def = ((defensive_rating['DFGM'].sum())/(defensive_rating['DFGA'].sum())*100)-defensive_rating['FG%'].mean()
```
